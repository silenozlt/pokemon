```python
import numpy as np 
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
pokemon = pd.read_csv("pokemon.csv")
```

```python
pokemon.describe()
```

```python
plt.hist(pokemon["type1"],alpha=1, label='Value', color="b",edgecolor="y", linewidth=2,bins=50)
plt.legend()
plt.ylabel('Numbers of Type', size='x-large')
plt.xlabel('Types Name', size= 'x-large')
plt.show()
```

```python
sns.barplot(x="is_legendary",y="speed",data=pokemon)
plt.legend()
plt.title("Speed Analysis of Pokemon", size='x-large')
plt.ylabel("Speed of Pokemon", size='large')
plt.xlabel("Legendary of Pokemon", size= 'large')
plt.show()
```

```python
sns.barplot(x="is_legendary",y="weight_kg",data=pokemon)
plt.legend()
plt.title("Weight Analysis of Pokemon", size='x-large')
plt.ylabel("Weight of Pokemon", size='large')
plt.xlabel("Legendary of Pokemon", size= 'large')
plt.show()
```

```python
sns.barplot(x="is_legendary",y="speed",palette="rocket",data=pokemon)
plt.legend()
plt.title("7 Generation Analysis of Pokemon", size='x-large')
plt.ylabel('Speed', size='large')
plt.xlabel('Legenday', size= 'large')
plt.show()
```

```python
sns.barplot(x="is_legendary",y="speed",palette="vlag",data=pokemon) #palette is a colour combination 
plt.legend()
plt.title("7 Generation Analysis of Pokemon", size='x-large')
plt.ylabel('Speed', size='large')
plt.xlabel('Legenday', size= 'large')
plt.show()
```

```python
sns.barplot(x="is_legendary",y="speed",hue="generation",data=pokemon)
plt.legend()
plt.title("7 Generation Analysis of Pokemon", size='x-large')
plt.ylabel('Speed', size='large')
plt.xlabel('Legenday', size= 'large')
plt.show()
```
